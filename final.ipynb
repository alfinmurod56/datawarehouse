```python
#Mengisi kolom yang memiliki nilai Null Value
df['bidikmisi'].fillna("non-bidikmisi", inplace=True)
df['Pilihan 2'].fillna("Belum Memilih", inplace=True)
df['Lulus pada Prodi'].fillna("Tidak Lulus", inplace=True)
df['Lulus Pilihan'].fillna("Tidak Lulus", inplace=True)
```

```python
#Melakukan drop
df=df.drop(['nomor_pendaftaran','Sekolah','Kabupaten','Pilihan 1','Pilihan 2','Predikat Semester 1','Predikat Semester 2'], axis=1)
```

##FULL

**df_full adalah dataset yang mencakup seluruh peserta baik yang lulus maupun tidak**

```python
df_full= df.fillna("0")
df_full['IP Sem 1'] = df_full['IP Sem 1'].astype(float)
df_full['IP Sem 2'] = df_full['IP Sem 2'].astype(float)
```

```python
#Menghitung nilai rata-rata IP
df_full['Rata_IP']=(df_full['IP Sem 1'] + df_full['IP Sem 2'])/2

#Memberikat ket predikat kepada rata2 IP
def grading(Rata_IP):
    if Rata_IP == 0 and Rata_IP <= 1:
        return "Sangat Kurang"
    elif Rata_IP > 1 and Rata_IP <= 2:
        return "Kurang"
    elif Rata_IP > 2 and Rata_IP <= 3:
        return "Baik"    
    else:
        return "Sangat Baik"

#Mengubah Lulus Pilihan dari string ke numerical
def Pil_lulus(lulus):
    if lulus == 'pil_1':
        return "2"
    elif lulus == 'pil_2':
        return "1"
    else:
      return "0"

df_full["Predikat"] = df_full['Rata_IP'].apply(grading)
df_full['Lulus Pilihan']=df_full['Lulus Pilihan'].apply(Pil_lulus)
df_full
```

```python
#df_full_1 adalah query hapus IP Sems 1&2, Rata IP
df_full_1=df_full.drop(['IP Sem 1','IP Sem 2','Rata_IP'], axis=1)
df_full_1
```

```python
#Menghitung nilai rata-rata IP
df_IP['Rata_IP']=(df_IP['IP Sem 1'] + df_IP['IP Sem 2'])/2

#Memberikan ket predikat kepada rata2 IP
def grading(Rata_IP):
    if Rata_IP == 0 and Rata_IP <= 1:
        return "Sangat Kurang"
    elif Rata_IP > 1 and Rata_IP <= 2:
        return "Kurang"
    elif Rata_IP > 2 and Rata_IP <= 3:
        return "Baik"    
    else:
        return "Sangat Baik"

#Mengubah Lulus Pilihan dari string ke numerical
def Pil_lulus(lulus):
    if lulus == 'pil_1':
        return "2"
    elif lulus == 'pil_2':
        return "1"
    else:
      return "0"

df_IP["Predikat"] = df_IP['Rata_IP'].apply(grading)
df_IP['Lulus Pilihan']=df_IP['Lulus Pilihan'].apply(Pil_lulus)
df_IP
```

##Encode

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
columns=['JK','bidikmisi','Predikat']
for column in columns:
  df_full_1[column]= label_encoder.fit_transform(df_full_1[column])
  df_IP_1[column]= label_encoder.fit_transform(df_IP_1[column])
```

##OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder


encoder = OneHotEncoder()

columns_to_encode = ['Provinsi', 'Lulus pada Prodi']


encoded_data_full_1 = encoder.fit_transform(df_full_1[columns_to_encode])
df_encoded_full_1 = pd.DataFrame(encoded_data_full_1.toarray())
encoded_columns_full_1 = encoder.get_feature_names_out(columns_to_encode)
df_encoded_full_1.columns = encoded_columns_full_1
df_full_1_encoded = pd.concat([df_full_1, df_encoded_full_1], axis=1)

encoded_data_IP_1 = encoder.transform(df_IP_1[columns_to_encode])
df_encoded_IP_1 = pd.DataFrame(encoded_data_IP_1.toarray())
encoded_columns_IP_1 = encoder.get_feature_names_out(columns_to_encode)
df_encoded_IP_1.columns = encoded_columns_IP_1
df_IP_1_encoded = pd.concat([df_IP_1, df_encoded_IP_1], axis=1)

print("df_full_1_encoded:")
print(df_full_1_encoded.head())
print("\ndf_IP_1_encoded:")
print(df_IP_1_encoded.head())

# Kalau pake ini hasilnya 0.0 atau 0.1
```
